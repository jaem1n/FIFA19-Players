# 가성비 선수와 연봉에 거품 낀 선수 찾기
포지션 그룹별로 다른 방법을 적용해 보았다.

## 수비수 & 골키퍼

1. Value와 Wage의 회귀선을 찾는다.
2. 회귀식을 따르는 값을 구한다.

#### 가성비
3. Value(시장가치)가 평균 이상인 선수만 남긴다.
4. 각 선수의 Wage에서 2번 값을 뺀 값를 구한다.
5. 3번 값이 가장 작은 음수인 선수를 구한다. 

#### 거품
3. Value(시장가치)가 평균 이하인 선수만 남긴다.
4. 각 선수의 Wage에서 2번 값을 뺀 값를 구한다.
5. 3번 값이 가장 큰 양수인 선수를 구한다.

### 가성비 높은 수비수 & 골키퍼

```python
def defender(position):
    df = fifa[fifa['Position']==position]
    model = LinearRegression()
    x = df[['Value']]
    y = df['Wage']
    model.fit(x, y)
    coef = model.coef_
    intercept = model.intercept_ 
    df['regression'] = coef * df['Value'] + intercept
    df = df[df['Value'] >= df['Value'].mean()]
    df['diff'] = df['Wage'] - df['regression']
    return df[df['diff'] == df['diff'].min()]
```

### 연봉에 거품 낀 수비수 & 골키퍼

```python
def defender2(position):
    df = fifa[fifa['Position']==position]
    model = LinearRegression()
    x = df[['Value']]
    y = df['Wage']
    model.fit(x, y)
    coef = model.coef_
    intercept = model.intercept_ 
    df['regression'] = coef * df['Value'] + intercept
    df = df[df['Value'] <= df['Value'].mean()]
    df['diff'] = df['Wage'] - df['regression']
    return df[df['diff'] == df['diff'].max()]
```

## 공격수

1. Potential(가능성)에서 Overall(현재 전반적 점수)을 뺀 값을 구해 Growth(성장가능성) 점수 컬럼을 만든다.

#### 가성비

2. Growth가 11점 이상인 선수만 남긴다.
3. Potential이 80점 이상인 선수만 남긴다. 
4. Wage(주급)가 평균 이하인 선수만 남긴다.

#### 거품

2. Growth가 0점으로 성장 가능성이 없는 선수만 남긴다.
3. Potential이 70점 이하인 선수만 남긴다.
4. Wage가 평균 이상인 선수만 남긴다.

```python
fifa['difference'] = fifa['Potential'] - fifa['Overall']
```

```python
def growth(d):
    if d==0:
        return "Maintain"
    elif d>=1 and d<=5:
        return "Small"
    elif d>=6 and d<=10:
        return "Medium"
    elif d>=11:
        return "Big"
```

```python
fifa['growth'] = fifa['difference'].apply(growth)
```

### 가성비 높은 공격수

```python
def forward(position):
    df = fifa[(fifa['Position']==position) & (fifa['growth']=='Big') & (fifa['Potential']>=80)]
    return df[df['Wage']<=df['Wage'].mean()]
```

```python
print('ST(Striker):', len(forward('ST')))
print('LS(Left Striker):', len(forward('LS')))
print('RS(Right Striker):', len(forward('RS')))
print('CF(Center Forward):', len(forward('CF')))
print('LF(Left Forward):', len(forward('LF')))
print('RF(Right Forward):', len(forward('RF')))
```

## 공격형 미드필더

1. Potential(가능성)에서 Overall(현재 전반적 점수)을 뺀 값을 구해 Growth(성장가능성) 점수 컬럼을 만든다.

#### 가성비

2. Growth가 19점 이상인 선수만 남긴다.
3. Wage per Overall(점수 1점당 주급)이 평균 이하인 선수만 남긴다.

#### 거품

2. Growth가 0점으로 성장 가능성이 없는 선수만 남긴다.
3. Wage per Overall이 평균 이상인 선수만 남긴다.

```python
def growth2(d):
    if d==0:
        return "Maintain"
    elif d>=1 and d<=9:
        return "Small"
    elif d>=10 and d<=18:
        return "Medium"
    elif d>=19:
        return "Big"
```

```python
fifa['growth2'] = fifa['difference'].apply(growth2)
```

```python
fifa['wage_per_overall'] = fifa['Wage'] / fifa['Overall']
```

```python
overall_wage_by_position = pd.DataFrame(fifa.groupby('Position')['wage_per_overall'].mean().astype(float))
overall_wage_by_position
```

### 가성비 높은 공격형 미드필더

```python
def att_mf(position):
    df = fifa[(fifa['Position']==position) & (fifa['growth2']=='Big')]
    return df[df['wage_per_overall']<=overall_wage_by_position.loc[position, 'wage_per_overall']]
```

```python
print('CAM(Center Attacking Midfielder):', len(att_mf('CAM')))
print('LAM(Left Attacking Midfielder):', len(att_mf('LAM')))
print('RAM(Right Attacking Midfielder):', len(att_mf('RAM')))
print('LW(Left Wing):', len(att_mf('LW')))
print('RW(Right Wing):', len(att_mf('RW')))
```

### 연봉에 거품 낀 공격형 미드필더

```python
def att_mf2(position):
    df = fifa[(fifa['Position']==position) & (fifa['growth2']=='Maintain')]
    return df[df['wage_per_overall']>=overall_wage_by_position.loc[position, 'wage_per_overall']]
```

```python
print('CAM(Center Attacking Midfielder):', len(att_mf2('CAM')))
print('LAM(Left Attacking Midfielder):', len(att_mf2('LAM')))
print('RAM(Right Attacking Midfielder):', len(att_mf2('RAM')))
print('LW(Left Wing):', len(att_mf2('LW')))
print('RW(Right Wing):', len(att_mf2('RW')))
```

## 중앙 미드필더

1. Potential(가능성)에서 Overall(현재 전반적 점수)을 뺀 값을 구해 Growth(성장가능성) 점수 컬럼을 만든다.

#### 가성비

2. Growth가 19점 이상인 선수만 남긴다.
3. Wage per Overall(점수 1점당 주급)이 평균 이하인 선수만 남긴다.
4. Release Clause(현 소속 구단과의 협의 없이 바로 협상 가능한 이적료 하한선)이 평균 이하인 선수만 남긴다.

#### 거품

2. Growth가 0점으로 성장 가능성이 없는 선수만 남긴다.
3. Wage per Overall이 평균 이상인 선수만 남긴다.
4. Release Clause이 평균 이상인 선수만 남긴다.

### 가성비 높은 중앙 미드필더

```python
def cen_mf(position):
    df = fifa[(fifa['Position']==position) & (fifa['growth2']=='Big')]
    return df[(df['wage_per_overall']<=overall_wage_by_position.loc[position, 'wage_per_overall']) & df['Release Clause']<=df['Release Clause'].mean()]
```

```python
print('CM(Central Midfielder):', len(cen_mf('CM')))
print('LCM(Left Central Midfielder):', len(cen_mf('LCM')))
print('RCM(Right Central Midfielder):', len(cen_mf('RCM')))
print('LM(Left Midfielder):', len(cen_mf('LM')))
print('RM(Right Midfielder):', len(cen_mf('RM')))
```

## 수비형 미드필더

1. 각 세부기술을 독립변수로, Potential을 종속변수로 ols 회귀모델에 넣는다.
2. R-squared 값이 0.25 이상인 기술만 남긴다.
    * Cohen (1992)
        * r <= 0.12 : Low
        * 0.12 < r <= 0.25 : Medium
        * r < 0.25 : High
3. Potential이 상위 10%인 선수만 남긴다.
4. 기술들 간 상관계수가 0.8 이상인 선수를 찾는다.
    * 잘하는 선수는 여러 기술점수가 대체적으로 다 높다는 것에서 착안
