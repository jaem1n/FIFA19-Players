포지션 점수
* ST=LS=RS
* CF=LF=RF
* CAM=LAM=RAM
* LW=RW
* LM=RM
* CM=LCM=RCM
* CDM=LDM=RDM
* LWB=RWB
* LB=RB
* CB=LCB=RCB

```python
def regressor(position):
    forest = RandomForestRegressor(random_state=42, n_estimators=2001)
    train_X = fifa[['Crossing', 'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling',
           'Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration',
           'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower',
           'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
           'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
           'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
           'GKKicking', 'GKPositioning', 'GKReflexes']]
    train_y = fifa[position]
    test_X = predict_position[['Crossing', 'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling',
           'Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration',
           'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower',
           'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
           'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
           'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
           'GKKicking', 'GKPositioning', 'GKReflexes']]
    test_y = predict_position[position]
    forest = forest.fit(train_X, train_y)
    predict_forest = forest.predict(test_X)
    predict_position[position] = predict_forest
    return predict_position
```

```python
# ST=LS=RS
regressor('ST')
```

```python
# LW=RW
regressor('LW')
predict_position['RW'] = predict_position['LW']
```

```python
# LM=RM
regressor('LM')
predict_position['RM'] = predict_position['LM']
```

```python
# CM=LCM=RCM
regressor('CM')
predict_position['LCM'] = predict_position['CM']
predict_position['RCM'] = predict_position['CM']
```

```python
# CDM=LDM=RDM
regressor('CDM')
predict_position['LDM'] = predict_position['CDM']
predict_position['RDM'] = predict_position['CDM']
```

```python
# LWB=RWB
regressor('LWB')
predict_position['RWB'] = predict_position['LWB']
```

```python
# LB=RB
regressor('LB')
predict_position['RB'] = predict_position['LB']
```

```python
# CB=LCB=RCB
regressor('CB')
predict_position['LCB'] = predict_position['CB']
predict_position['RCB'] = predict_position['CB']
```
